```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\GPT BANTWAL\\Documents\\credit_risk.csv')
df
```

```python
x=df.drop(['class'],axis=1)
x
```

```python
y=df['class']
y
```

```python
df_encoded=pd.get_dummies(x)
df_encoded
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df_encoded,y,test_size=0.15,random_state=100)
```

```python
xtrain.shape
xtest.shape
ytrain.shape
ytest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=1)
model.fit(xtrain,ytrain)
```

```python
train_predictions=model.predict(xtrain)
train_predictions
```

```python
test_predictions=model.predict(xtest)
test_predictions
```

```python
train_accuracy=model.score(xtrain,ytrain)
train_accuracy
```

```python
train_accuracy=model.score(xtest,ytest)
train_accuracy
```
